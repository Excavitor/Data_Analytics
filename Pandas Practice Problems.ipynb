# Problem 2

```python
series = pd.Series(['01 Jan 2015', '10-02-2016', '20180307', '2014/05/06', '2016-04-12', '2019-04-06T11:20'])

series = series.map(lambda x: parse(x))

month = series.dt.day.tolist()
year = series.dt.dayofyear.tolist()
week_number = series.dt.isocalendar().week.tolist()
week = series.dt.strftime('%A').tolist()

print("Original Series:\n")
print(series)
print("\nDay of month:")
print(month)
print("\nDay of year:")
print(year)
print("\nWeek number:")
print(week_number)
print("\nDay of week:")
print(week)
```

# Problem 4

```python
list_1 = {'student_id': ['S1', 'S2', 'S3', 'S4', 'S5'],
         'name': ['Danniella Fenton', 'Ryder Storey', 'Bryce Jensen', 'Ed Bernal', 'Kwame Morin'],
         'marks': [200, 210, 190, 222, 199]}

list_2 = {'student_id': ['S4', 'S5', 'S6', 'S7', 'S8'],
         'name': ['Scarlette Fisher', 'Carla Williamson', 'Dante Morse', 'Kaiser William', 'Madeeha Preston'],
         'marks': [201, 200, 198, 219, 201]}

list_3 = {'student_id': ['S1', 'S2', 'S3', 'S4', 'S5', 'S7', 'S8', 'S9', 'S10', 'S11', 'S12', 'S13'],
         'exam_id': [23, 45, 12, 67, 21, 55, 33, 14, 56, 83, 88, 12]}


student_data1 = pd.DataFrame(list_1)
student_data2 = pd.DataFrame(list_2)
exam_data = pd.DataFrame(list_3)


result_1 = pd.concat([student_data1, student_data2])
result_2 = pd.merge(result_1, exam_data, on='student_id')


print("student_data1 and student_data2 join\n")
print(result_1)
print("\njoin along student_id\n")
print(result_2)
```
