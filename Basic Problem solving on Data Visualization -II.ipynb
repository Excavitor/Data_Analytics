# Practice problem 6.7
Compare the total sales price in year-wise quarterly between Dhaka and Chittagong, and plot it into a Scatter Chart.

```python
add_customer_time_with_fact = pd.merge(pd.merge(fact_table, time_dim, on='time_key'), customer_dim, on='coustomer_key')
add_customer_time_with_fact
```

```python
quarter_wise_div = add_customer_time_with_fact[(add_customer_time_with_fact['division'] == 'Dhaka') | (add_customer_time_with_fact['division'] == 'Chittagong')]
quarter_wise_div
```

```python
total_sales_per_quarter = quarter_wise_div.groupby(['year', 'quarter', 'division'])['total_price'].sum().reset_index()
total_sales_per_quarter
```

```python
plt.figure(figsize=(12, 8))
sns.set_style('darkgrid')
sns.scatterplot(data=total_sales_per_quarter, x='quarter', y='total_price', hue='division')
plt.title('Total Sales Price per Quarter')
plt.xlabel('Quarter')
plt.ylabel('Total Sales Price')
plt.show()
```

# Practice problem 6.9
Break down year-wise the total sales geographically (division-wise) and plot it into Stripplot.

```python
add_customer_time_with_fact_6_9 = pd.merge(pd.merge(fact_table, time_dim, on='time_key'), customer_dim, on='coustomer_key')
add_customer_time_with_fact_6_9
```

```python
total_sales_per_year = add_customer_time_with_fact_6_9.groupby(['year', 'division'])['total_price'].sum().reset_index()
total_sales_per_year
```

```python
plt.figure(figsize=(12, 8))
sns.set_style('darkgrid')
sns.stripplot(data=total_sales_per_year, x='year', y='total_price', hue='division')
plt.title('Year and Division')
plt.xlabel('Year')
plt.ylabel('Total Sales Price')
plt.legend(title='Division')
plt.show();
```
